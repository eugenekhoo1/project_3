# Problem Statement

This project aims to classify a user's post into one of two categories - Anxiety, Depression. With this information, moderators on forums (without labelled data) are able to identify at risk users and allocate the appropriate help/resources to them.

# Import Packages / Prepare Data

```python
# Import packages

import pandas as pd
import string
import re
import nltk
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Import scraped data

df_anx = pd.read_csv('data/anxiety.csv')
df_dep = pd.read_csv('data/depression.csv')
```

```python
# Drop rows with null selftext

df_anx.drop(df_anx[df_anx['selftext'].isnull() == True].index, inplace=True)
df_dep.drop(df_dep[df_dep['selftext'].isnull() == True].index, inplace=True)
```

```python
# Drop rows without selftext

df_anx.drop(df_anx[df_anx['selftext'] == '[removed]'].index, inplace=True)
df_dep.drop(df_dep[df_dep['selftext'] == '[removed]'].index, inplace=True)
```

## Preparing selftext for analysis

```python
# Create list of punctuations to remove
# Create list of stopwords using NLTK

punc_str = string.punctuation + "â€™"
stopword = nltk.corpus.stopwords.words('english')
```

```python
# Clean text
def clean_text(text):
    # remove punctuations
    text_ = str(text).lower()
    text_nopunc = "".join([char for char in text_ if char not in punc_str])
    
    # tokenize
    text_nopunc = str(text_nopunc)
    tokenized = re.split('\W+', text_nopunc)
    
    return tokenized
```

```python
df_anx['selftext_tokenized'] = df_anx['selftext'].apply(lambda x: clean_text(x))
df_dep['selftext_tokenized'] = df_dep['selftext'].apply(lambda x: clean_text(x))
```

```python
# Remove stopwords

def remove_stop(text):
    text_nostop = [word for word in text if word not in stopword]
    return text_nostop
```

```python
df_anx['selftext_nostop'] = df_anx['selftext_tokenized'].apply(lambda x: remove_stop(x))
df_dep['selftext_nostop'] = df_dep['selftext_tokenized'].apply(lambda x: remove_stop(x))
```

```python
# Create lemmatizing object

wnl = nltk.WordNetLemmatizer()
```

```python
# Lemmatizer function

def lemmatize(tokenized_text):
    text = [wnl.lemmatize(word) for word in tokenized_text]
    return text
```

```python
df_anx['selftext_lemmatized'] = df_anx['selftext_nostop'].apply(lambda x: lemmatize(x))
df_dep['selftext_lemmatized'] = df_dep['selftext_nostop'].apply(lambda x: lemmatize(x))
```

```python
# Remove words with numbers

def remove_number(text):
    text = [re.sub(r"^\d+|\d$","",word) for word in text]
    return text
```

```python
df_anx['selftext_clean'] = df_anx['selftext_lemmatized'].apply(lambda x: remove_number(x))
df_dep['selftext_clean'] = df_dep['selftext_lemmatized'].apply(lambda x: remove_number(x))
```

```python
# Rejoin tokens to string

def join_tokens(text):
    text = " ".join([char for char in text])
    return text
```

```python
df_anx['selftext_clean_string'] = df_anx['selftext_clean'].apply(lambda x: join_tokens(x))
df_dep['selftext_clean_string'] = df_dep['selftext_clean'].apply(lambda x: join_tokens(x))
```

# Length of Posts

```python
# String Lengths

df_anx['string_length'] = df_anx['selftext'].map(len)
df_dep['string_length'] = df_dep['selftext'].map(len)
```

```python
plt.figure(figsize=(15,8))
plt.title('Distribution of String Lengths', size=16)
plt.xlabel('String Length')
plt.ylabel('Probability Density')
plt.hist(df_anx['string_length'], bins=100, alpha=0.3, color='b', density=True)
plt.hist(df_dep['string_length'], bins=100, alpha=0.5, color='y', density=True)
plt.show()
```

```python
# Word Lenghts

df_anx['word_length'] = df_anx['selftext'].map(lambda x: len(x.split()))
df_dep['word_length'] = df_dep['selftext'].map(lambda x: len(x.split()))
```

```python
plt.figure(figsize=(15,8))
plt.title('Distribution of Word Lengths', size=16)
plt.xlabel('Word Length')
plt.ylabel('Probability Density')
plt.hist(df_anx['word_length'], bins=100, alpha=0.3, color='b', density=True)
plt.hist(df_dep['word_length'], bins=100, alpha=0.5, color='y', density=True)
plt.show()
```

Distribution of string and word lengths are roughly the same, leaving us without much to infer from

# Word Frequency

```python
def word_freq(n, data):
    # n : n-gram
    # data : series of strings
    
    cv = CountVectorizer(ngram_range=(n,n))
    cv_anx = cv.fit_transform(data['selftext_clean_string'])
    df_anx = pd.DataFrame(cv_anx.todense(), columns=cv.get_feature_names_out())
    
    freq_data = df_anx.sum().sort_values(ascending=False).head(20)
    
    plt.figure(figsize=(15,8))
    plt.title(f"Frequency for {n}-Word Length", size=16)
    plt.xlabel('Frequency')
    plt.barh(freq_data.index, freq_data.values)
```
